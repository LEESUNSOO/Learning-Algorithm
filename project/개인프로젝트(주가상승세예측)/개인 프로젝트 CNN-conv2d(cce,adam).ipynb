```python
import pandas as pd
import numpy as np

data2 = pd.read_excel('모델인풋데이터셋처리전DF2.xls')

data2.drop(columns='Unnamed: 0', inplace=True)
data2.dropna(inplace=True)
dataT = data2.T

dataT
```

```python
#2차원배열 형태의 데이터셋 만들기

def createData(df, look_back = 1):
  dataX, dataY = [], []
  length = df.shape[1] 
  for i in range(length - look_back):
    dataX.append(dataT.iloc[1:5, i:i+look_back].to_numpy())
    dataY.append(dataT.iloc[5,i+look_back])

  X = np.asarray(dataX, dtype=np.float32)
  Y = np.asarray(dataY, dtype=np.float32)
  return X, Y
```

```python
dataT.shape[1] 

#462의 길이

dataX, dataY = createData(dataT, 30)

#363 : 69 ( 8 : 2 비율로 train/test)

train_x = dataX[:362+1]
train_y = dataY[:362+1]
test_x = dataX[362+1:]
test_y = dataY[362+1:]
print(len(train_y))
print(len(test_x))
inf_row = 4
inf_col = 30

train_x
```
