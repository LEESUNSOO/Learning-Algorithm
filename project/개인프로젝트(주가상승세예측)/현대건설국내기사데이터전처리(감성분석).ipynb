```
from konlpy.tag import Kkma
from konlpy.utils import pprint
kkma = Kkma()
nltk.download('averaged_perceptron_tagger')
```

```
def tri_gram(pos_tag):
  tri_gram_list = []
  length = len(pos_tag)

  for idx in range(length-2):
    temp = ''
    temp = '{0}/{1};{2}/{3};{4}/{5}'.format(pos_tag[idx][0], pos_tag[idx][1], pos_tag[idx+1][0], pos_tag[idx+1][1],pos_tag[idx+2][0], pos_tag[idx+2][1])
    tri_gram_list.append(temp)

  return tri_gram_list
```

```
def bi_gram(pos_tag):
  bi_gram_list = []
  length = len(pos_tag)

  for idx in range(length-2):
    temp = ''
    temp = '{0}/{1};{2}/{3}'.format(pos_tag[idx][0], pos_tag[idx][1], pos_tag[idx+1][0], pos_tag[idx+1][1])
    bi_gram_list.append(temp)

  return bi_gram_list
```

```
def one_gram(pos_tag):
  one_gram_list = []
  length = len(pos_tag)

  for idx in range(length-2):
    temp = ''
    temp = '{0}/{1}'.format(pos_tag[idx][0], pos_tag[idx][1])
    one_gram_list.append(temp)

  return one_gram_list
```

```
dnewsDate = []
dnewsScore = []
for idx in range(229):
  score = 0
  text = df.iloc[idx]['기사내용']
  # if '배구' in text:
  #   score = 0
  # else:
  pos_tag = kkma.pos(text)
  result = tri_gram(pos_tag) + bi_gram(pos_tag) +one_gram(pos_tag)
  #print(result)
  #result.append('가*/JKS;있/VV')
  #print(result)
  for gram in result:
    if gram in lexi_dict[0].keys():
      #print(gram)
      #print(lexi_dict[0][gram])
      score += lexi_dict[0][gram]
  dnewsDate.append(df.iloc[idx]['Unnamed: 0'])
  dnewsScore.append(score)  
    
```

```
dnews = []
for d, s in zip(dnewsDate, dnewsScore):
  dnews.append([d,s])

dnews
```

```
dnews_dict = dict(dnews)
dnews_dict

temp = pd.Series(dnews_dict).to_frame()
temp
```
