```python
from konlpy.corpus import kolaw
data = kolaw.open('constitution.txt').read()

from konlpy.tag import Komoran
komoran = Komoran()

print(komoran.nouns("%r"%data[0:1000]))

word_list = komoran.nouns("%r"%data[0:1000])

text = ' '.join(word_list)

```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud

wordc = WordCloud()
wordc.generate(text)

plt.figure()
plt.imshow(wordc, interpolation="bilinear")

# wordc = WordCloud(background_color='white', max_words=20, font_path= '/home/esuns/Download/NanumGothic.ttf', relative_scaling=0.2)
wordc = WordCloud(background_color='white', max_words=20, relative_scaling=0.2)

wordc.generate(text)

plt.figure()
plt.imshow(wordc, interpolation='bilinear')
plt.axis('off')
```
