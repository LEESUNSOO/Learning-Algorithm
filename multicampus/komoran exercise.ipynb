```python
from konlpy.tag import Komoran
```

```python
komoran = Komoran()
```

```python
myhangul = u'아름답지만 다소 복잡하기도 한 한국어는 세계에서 13번째로 많이 사용하는 언어입니다.'
```

```python
komoran.morphs(myhangul)
```

```python
komoran.nouns(myhangul)
```
