```
import requests
from bs4 import BeautifulSoup as bs

url = 'http://www.koreatimes.co.kr/www/opinion/2019/09/202_275943.html'

r = requests.get(url)
r = r.text

soup = bs(r, "html.parser")
soup

s = soup.select_one('#startts > span')
print(s.text)
```

```
s= s.text
```

```
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
nltk.download('stopwords')
nltk.download('punkt')
stop_words = set(stopwords.words('english')) 

word_tokens = word_tokenize(s)
```

```
result = []
for w in word_tokens: 
  ls = []
  if w not in stop_words:
    ls.append(w)
    result.append(ls) 

# print(word_tokens) 
# print(result)
```

```

len(result)
result[:10]
```

```
from gensim.models import Word2Vec
model = Word2Vec(sentences=result, size=400, window=5, min_count=5, workers=4, sg=0)
```

```
print(model.most_similar(positive=["Trump"], topn=5))
```
